# 第一題、匯入資料集(17% = 4% + 4% + 9%)
#### (1)利用input指令讓使用者自行輸入已共享google試算表的連結代碼（變數名稱為ID）以取得後續資料
#### (2)將抓取google試算表的下載網址定義成一個字串變數(Filename)
#### (3)利用pandas中的read_excel分別讀入「Financial_Data」、「ESG_Score」、「OTHERS」三張頁籤的資料（工作表名稱即為資料集的變數名稱）

```python
import pandas as pd
```

```python
ID=input("請輸入google試算表的連結代碼:")
Filename="https://drive.google.com/uc?id="+ID+"&export=download"
Financial_Data=pd.read_excel(Filename,sheet_name="Financial_Data")
Financial_Data
```

```python
ESG_Score=pd.read_excel(Filename,sheet_name="ESG_Score")
ESG_Score
```

```python
OTHERS=pd.read_excel(Filename,sheet_name="OTHERS")
OTHERS
```

# 第二題、合併資料集（用pd.merge 或 pd.concat）(5%)
#### 請將上面三個資料集（Financial_Data、ESG_Score、OTHERS）合併成一個資料集(Company_ESG)

```python
merged1=pd.merge(Financial_Data, ESG_Score, on="CompanyID", how="inner")
Company_ESG = pd.merge(merged1, OTHERS, on="CompanyID",how="inner")
```

```python
Company_ESG = Company_ESG.rename(columns={'ESG_Overall_x': 'ESG_Overall', 'ESG_Environmental_x': 'ESG_Environmental', 'ESG_Social_x': 'ESG_Social','ESG_Governance_x':'ESG_Governance','CarbonEmissions_x':'CarbonEmissions','WaterUsage_x':'WaterUsage','EnergyConsumption_x':'EnergyConsumption'})
```

```python
Company_ESG
```

```python
Company_ESG = Company_ESG[[col for col in Company_ESG.columns if not col.endswith('_y')]]
Company_ESG.columns = [col.replace('_x', '') for col in Company_ESG.columns]
```

```python
Company_ESG.head()
```

# 第六題、資料排序(5%)
#### 請將「資料集Company_ESG」的資料依ESG_Overall大小由大至小排序，並將排序結果回存至「資料集Company_ESG」中

# 第七題、重新定義變數涵意及新增變數(18% = 3% + 10% + 5%)
#### (1)建立一個空的list變數(New_Group)
#### (2)利用for迴圈及ifelse判斷式將「資料集Company_ESG」中的ESG_Overall重新定義成3組別：
- 小於60分(不含60)
- 60-80分(不含80)
- 80分以上

#### (3)在「資料集Company_ESG」中新增一個欄位為「ESG成績分組」。

```python
New_Group=list()
for i in Company_ESG["ESG_Overall"]:
  if i<60:
    New_Group.append("0-60")
  elif i<80:
    New_Group.append("60-80")
  else:
    New_Group.append("80+")
Company_ESG["ESG成績分組"]=New_Group
```

```python
Company_ESG
```
